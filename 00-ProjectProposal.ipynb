## Research Question

Did majors with higher exposure to AI (conceptually defined by major’s focus domain and task structures) experience different trends in employment and graduate school enrollment after the rise of modern AI (Rise of Transformer after 2017)?


## Background and Prior Work

 As many college students approach graduation, the job market has raised growing concerns about employment prospects and the decision to pursue graduate education. The rise of modern artificial intelligence, particularly since the introduction of the Transformer architecture in 2017, has significantly reshaped both scientific research and labor market demands. Previous research has documented the uneven adoption of artificial intelligence across different academic fields. Hajkowicz et al. (2023) conducted a bibliometric analysis covering 333 research fields from 1960 to 2021 and reveals that AI adoption varies significantly across the physical sciences, natural sciences, life sciences, social sciences, and the arts and humanities (https://www.semanticscholar.com/paper/Artificial-intelligence-adoption-in-the-physical-A-Hajkowicz-Sanderson/4d0d70769451c0f4e702bf860da26d58fa33e81b). 
Thayyib et al. (2023) found that key AI technologies such as machine learning, deep learning, and natural language processing remain concentrated in specific domain clusters, while other fields like engineering and biomedical sciences show relatively limited AI application.
(https://www.semanticscholar.com/paper/State-of-the-Art-of-Artificial-Intelligence-and-Big-Thayyib-Mamilla/641b46fc44d6c7db82640ad6f15da260589513ef).
Additionally, Gao and Wang (2023) developed a measurement framework to quantify both the direct use and potential benefit of AI across scientific disciplines, showing that the benefits of AI vary across fields and may exacerbate existing inequalities in science (https://www.semanticscholar.com/paper/Quantifying-the-Benefit-of-Artificial-Intelligence-Gao-Wang/e2f04c45223ab68cfd0ec7a2f11a27347e3565e0).
While these studies provide valuable insights into AI's differential impact across fields, little research has directly examined how majors with different AI exposure have experienced different trends in employment and graduate school enrollment. This study aims to fill this gap by analyzing post-2017 trends, using data of UCSD students on unemployment rates, industry, and graduate year across different majors, to understand whether AI exposure has become a significant factor shaping undergraduate outcomes.

## Hypothesis


We hypothesize that majors with higher exposure to AI (e.g., Math, CS, DS) will have a different trend in employment and graduate school enrollment than majors with lower exposure to AI (e.g., Philosophy, Literature, Art) after the rise of Transformer. We think so because majors with higher exposure to AI are more likely to play significant roles in the rapid expansion of AI in various industries, therefore vastly impacting students’ future development.

A. What variables?

    Individual-level variables
        Graduation year
        Undergraduate major
        Whether the person is employed, unemployed, or out of the labor force
        Occupation
        Industry
        Whether the person is enrolled in graduate school
        Type of graduate program, such as master’s, PhD, law school, medical school, or professional degree
    AI-use variables
        Whether a person uses AI tools such as ChatGPT, Claude, Gemini, or Microsoft Copilot
        Frequency of AI use: daily, weekly, monthly, rarely, or never
    AI-exposure variables
        Industry-level AI adoption rate
        Firm-level AI adoption rate, if available
        State-level AI adoption rate, if available
        Whether the major leads to tasks that are highly exposed to AI, such as coding, writing, design, analysis, research, or administrative work
    Labor market control variables
        State-level monthly unemployment rate
        Occupation-level unemployment rate
        Industry-level unemployment rate
        State-level economic conditions
        Demographic controls, such as age, gender, race/ethnicity, citizenship status, and parental education if available


B. How many observations are needed?

I would want a large dataset with at least several thousand person-year observations.

The dataset should include multiple graduation cohorts from before and after 2017.

Ideally, it would cover years from around 2010 to the present, so there are enough years before and after the rise of transformer-based AI.

C. Who/what/how would these data be collected?

The data would ideally be collected from college graduates in the United States.

Universities could provide administrative data on major, graduation year, degree type, and graduate school enrollment.

Government labor surveys could provide employment status, occupation, industry, earnings, and location.

A separate survey could ask individuals whether they use AI tools such as ChatGPT, Claude, Gemini, or Copilot and how often they use them.

D. How would these data be stored/organized?

The ideal dataset would be stored in a person-year format. Each row would represent one person in one year.Each column would represent a variable, such as major, year, employment status, graduate school enrollment, occupation, industry, AI use, AI exposure, and unemployment rate.
